```python
import shutil
import glob
import os

# Create directory structure in dataset2
!mkdir -p /content/dataset2/train/images
!mkdir -p /content/dataset2/train/labels
!mkdir -p /content/dataset2/valid/images
!mkdir -p /content/dataset2/valid/labels
!mkdir -p /content/dataset2/test/images
!mkdir -p /content/dataset2/test/labels
```

```python
source_dir = '/content/drive/MyDrive/filtered_dataset'
target_dir = '/content/dataset2'

splits = ['train', 'valid', 'test']
classes = ['bike', 'bus', 'car', 'traffic_sign', 'traffic_light', 'truck']

for split in splits:
    for cls in classes:
        src_images = f'{source_dir}/{cls}/{split}/images/*'
        src_labels = f'{source_dir}/{cls}/{split}/labels/*'
        dst_images = f'{target_dir}/{split}/images/'
        dst_labels = f'{target_dir}/{split}/labels/'

        # Copy images
        for img in glob.glob(src_images):
            shutil.copy(img, dst_images)
        # Copy labels
        for lbl in glob.glob(src_labels):
            shutil.copy(lbl, dst_labels)
```

```python
with open('/content/dataset2/data.yaml', 'w') as f:
    f.write('train: /content/dataset2/train/images\n')
    f.write('val: /content/dataset2/valid/images\n')
    f.write('test: /content/dataset2/test/images\n')
    f.write('nc: 6\n')  # Number of classes
    f.write("names: ['bike', 'bus', 'car', 'traffic_sign', 'traffic_light', 'truck']\n")  # Class names with IDs 0-5
```

```python
from ultralytics import YOLO
model = YOLO('/content/drive/MyDrive/yolo11m.pt')
```

```python
model.train(
    data='/content/dataset2/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='yolov11m_custom'
)
```

```python
model.val()
```
